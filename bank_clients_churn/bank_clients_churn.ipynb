## Исследование задачи

```python
def plot_roc_auc(model, features, target): #функция построение roc - кривых и подсчета roc-auc

    probabilities_features= model.predict_proba(features)
    probabilities_one_features = probabilities_features[:, 1]

    fpr, tpr, thresholds = roc_curve(target, probabilities_one_features)
    
    plt.figure(figsize=(5, 5))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая")
    plt.show()

    auc_roc = roc_auc_score(target, probabilities_features[:, 1])
    print('AUC-ROC:', auc_roc)
```

```python
def importances(model):   #функция определеия метрики важности признаков и построения столбчатого графика их значений 
    importances = model.feature_importances_
    pd_importances = pd.Series(importances, features.columns).sort_values(ascending=False)
    plt.figure(figsize=(15, 5))
    plt.bar(x=pd_importances.index, height=pd_importances.values)
    plt.xticks(rotation=90)
    for i in range(len(pd_importances.index)):
        plt.text(x = i, y = pd_importances.values[i], s = pd_importances.values[i].round(3), ha = 'center')
    plt.ylabel('Mean decrease in impurity')
    plt.title('Feature importances')
    plt.show()
```

### Изучение моделей без учета дисбаланса классов

Исследуем модели обучения (линейную регресиию, решающее дерево и случайный лес) без учета дисбаланса классов. Для каждой модели классификации определим ее метрики (recall, precision, f1).  Найдем наилучшую модель с оптимальными гиперпараметрами и исследуем её на адекватность.

```python
model_lr = LogisticRegression(random_state=12345, solver='liblinear')# исследование модели логистической регрессии
model_lr.fit(features_train, target_train)
predicted_valid_lr = model_lr.predict(features_valid) # предсказание на валидационной выборке 
recall_score_lr = recall_score(target_valid, predicted_valid_lr)#блок расчета метрик классификации 
precision_score_lr = precision_score(target_valid, predicted_valid_lr)
f1_score_lr = f1_score(target_valid, predicted_valid_lr)

print('LogisticRegression')
print('recall_score:', recall_score_lr)
print('precision_score:', precision_score_lr)
print('f1_score:', f1_score_lr)
print()
print('confusion_matrix:')
print(confusion_matrix(target_valid, predicted_valid_lr))
```

```python
plot_roc_auc(model_lr, features_valid, target_valid)
```

```python
max_f1_dt = 0 # исследование модели обучающего дерева
best_recall_score_dt = 0
best_precision_score_dt = 0
best_confusion_matrix_dt = 0
best_depth_dt = 0
for depth in range(1, 100):
    model_dt = DecisionTreeClassifier(random_state=12345, max_depth=depth) # создание модeли решающего дерева с заданной глубиной дерева
    model_dt.fit(features_train, target_train) # обучение модели
    predicted_valid_dt = model_dt.predict(features_valid) #блок расчета метрик классификации
    recall_score_dt = recall_score(target_valid, predicted_valid_dt)
    precision_score_dt = precision_score(target_valid, predicted_valid_dt)
    f1_score_dt = f1_score(target_valid, predicted_valid_dt)
    conf_matrix_dt = confusion_matrix(target_valid, predicted_valid_dt)
    if f1_score_dt > max_f1_dt:
        max_f1_dt = f1_score_dt
        best_depth_dt = depth
        best_recall_score_dt = recall_score_dt
        best_precision_score_dt = precision_score_dt
        best_confusion_matrix_dt = conf_matrix_dt

print('DecisionTreeClassifier')
print('depth:', best_depth_dt)
print('recall_score:', best_recall_score_dt)
print('precision_score:',  best_precision_score_dt)
print('max_f1_score:', max_f1_dt)
print()
print('confusion_matrix:')
print(best_confusion_matrix_dt)
```

## Борьба с дисбалансом

Для повышения качества прогнозирования и увеличения f1 метрики проверим несколько техник балансировки классов на полученной модели случайного леса с учетом найденных оптимальных значений гипперпараметров. Кроме того, для оценки качества исследуемых моделей построим для каждой из них 'ROC'-кривые и найдем площади по этим кривыми (AUC-ROC).

Для автоматизации построения 'ROC'-кривых и подсчета AUC-ROС создадим специализированную функцию.

```python
def plot_roc_auc(model, features, target):

    probabilities_features= model.predict_proba(features)
    probabilities_one_features = probabilities_features[:, 1]

    fpr, tpr, thresholds = roc_curve(target, probabilities_one_features)
    
    plt.figure(figsize=(5, 5))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая")
    plt.show()

    auc_roc = roc_auc_score(target, probabilities_features[:, 1])
    print('AUC-ROC:', auc_roc)
```

### Увеличение выборки

Приведение классов к балансу методом увеличения выборок предполагает достижения баланса примерно 1/1. В целях получения наиболее высокой метрики f1 проведем подбор коэффициента кратности увеличения значений положительного класса целевого признака и соответствующих ему признаков и определим оптимальный при этом баланс.

```python
#функция получения выборок таблицы признаков и целевого признака с увеличением положительного класса 
def upsample(features, target, repeat): 
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

best_repeat_qnty = 0
best_f1_rf_ups = 0
best_recall_score_rf_ups = 0
best_precision_score_rf_ups = 0
best_confusion_matrix_rf_ups = 0
for repeat in range (1, 10): #цикл подбора кратности увеличения выборок с учетом лучшей метрики  f1
    features_upsampled, target_upsampled = upsample(features_train, target_train, repeat)
    model_rf_ups = RandomForestClassifier(random_state=12345, n_estimators=best_n_estimators_rf, 
                                          max_depth=best_depth_rf)
    model_rf_ups.fit(features_upsampled, target_upsampled)
    predicted_valid_rf_ups = model_rf_ups.predict(features_valid)
    recall_score_rf_ups = recall_score(target_valid, predicted_valid_rf_ups)
    precision_score_rf_ups = precision_score(target_valid, predicted_valid_rf_ups)
    conf_matrix_rf_ups = confusion_matrix(target_valid, predicted_valid_rf_ups)
    f1_rf_ups = f1_score(target_valid, predicted_valid_rf_ups)
    if f1_rf_ups > best_f1_rf_ups:
        best_recall_score_rf_ups = recall_score_rf_ups
        best_precision_score_rf_ups = precision_score_rf_ups
        best_confusion_matrix_rf_ups = conf_matrix_rf_ups
        best_f1_rf_ups = f1_rf_ups
        best_repeat_qnty = repeat
        a = target_upsampled.value_counts(normalize=True)

        
print('recall_score:', best_recall_score_rf_ups)
print('precision_score:',  best_precision_score_rf_ups)
print('confusion_matrix:')
print(best_confusion_matrix_rf_ups)
print()
print('F1:', best_f1_rf_ups) #печать результатов
print('repeats:', best_repeat_qnty)
print()
print('class_balance:') #оптимальный баланс классов
print(a)
```

### Уменьшение выборки

Аналогично для метода уменьшения выборок в целях получения наиболее высокой метрики f1 проведем подбор коэффициента кратности уменьшения значений отрицательного класса целевого признака и соответсвующих ему признаков и определим оптимальный при этом баланс.

```python
#функция получения выборок таблицы признаков и целевого признака с уменьшение отрицательного класса 
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

best_fraction = 0
best_f1_rf_dwns = 0
best_recall_score_rf_dwns = 0
best_precision_score_rf_dwns = 0
best_confusion_matrix_rf_dwns = 0
for fraction in np.arange(0.1, 0.9, 0.1): #цикл подбора кратности уменьшения выборок с учетом лучшей метрики  f1
    features_downsampled, target_downsampled = downsample(features_train, target_train, fraction)
    model_rf_dwns = RandomForestClassifier(random_state=12345, n_estimators=best_n_estimators_rf, 
                                           max_depth=best_depth_rf)
    model_rf_dwns.fit(features_downsampled, target_downsampled)
    predicted_valid_rf_dwns = model_rf_dwns.predict(features_valid)
    recall_score_rf_dwns = recall_score(target_valid, predicted_valid_rf_dwns)
    precision_score_rf_dwns = precision_score(target_valid, predicted_valid_rf_dwns)
    conf_matrix_rf_dwns = confusion_matrix(target_valid, predicted_valid_rf_dwns)
    f1_rf_dwns = f1_score(target_valid, predicted_valid_rf_dwns)
    if f1_rf_dwns > best_f1_rf_dwns:
        best_recall_score_rf_dwns = recall_score_rf_dwns
        best_precision_score_rf_dwns = precision_score_rf_dwns
        best_confusion_matrix_rf_dwns = conf_matrix_rf_dwns
        best_f1_rf_dwns = f1_rf_dwns
        best_fraction = fraction
        b = target_downsampled.value_counts(normalize=True)
        
print('recall_score:', best_recall_score_rf_dwns)
print('precision_score:',  best_precision_score_rf_dwns)
print('confusion_matrix:')
print(best_confusion_matrix_rf_dwns)
print()
print('F1:', best_f1_rf_dwns)#печать результатов
print('fraction:', best_fraction)
print()
print('class_balance:')#оптимальный баланс классов
print(b)
```

Метрика f1 стала выше. Оптимальный баланс классов - примерно 2/1 - аналогичен для метода увеличения выборки. ROC-кривая исследуемой модели с уменьшенной выборкой находится примерно на том же уровне, что и для модели с увеличенной выборкой, значение AUC-ROC стало выше в четвертом знаке после запятой.
